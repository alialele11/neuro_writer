```python
import numpy as np
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tok = GPT2Tokenizer.from_pretrained("/content/drive/MyDrive/model/essays/checkpoint-5500")
model = GPT2LMHeadModel.from_pretrained("/content/drive/MyDrive/model/essays/checkpoint-5500")
```

```python
text = "Если звёзды зажигаются,"
inpt = tok.encode(text, return_tensors="pt")
```

```python
import random
k = random.randrange(1, 20)
t = float(random.randrange(1, 50))
out = model.generate(inpt, max_length=200, repetition_penalty=3.0, do_sample=True, top_k=k, top_p=0.95, temperature=t)
```
